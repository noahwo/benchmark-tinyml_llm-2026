# Bernoulli / Binomial Probability Calculator

This notebook computes the probability of getting exactly `x` successes in `n` independent Bernoulli trials with success probability `p` per trial:

$P(X=x)=\binom{n}{x}p^x(1-p)^{n-x}$

```python
import math

def bernoulli_trials_probability(n: int, p: float, x: int) -> float:
    """Return P(X=x) where X ~ Binomial(n, p)."""
    if n < 0:
        raise ValueError('n must be a non-negative integer')
    if not (0.0 <= p <= 1.0):
        raise ValueError('p must be between 0 and 1')
    if x < 0 or x > n:
        raise ValueError('x must be an integer in [0, n]')

    return math.comb(n, x) * (p ** x) * ((1.0 - p) ** (n - x))
```

```python
 
arg_x=[1]

arg_p=0.3
arg_n=25

prob=0.0
for x in arg_x:
    prob+=bernoulli_trials_probability(arg_n, arg_p, x)
    
print(f"P(X in {arg_x}) with n={arg_n}, p={arg_p} is {prob:.12g}")

 
```
